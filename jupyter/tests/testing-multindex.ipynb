```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.random((10, 8)))

level0 = ['Total Time including setup', 'Total Alignment time']
level1 = ['classic', 'classic-sp', 'recomp', 'recomp-sp']

index = pd.MultiIndex.from_product([level0, level1])
df.columns = index

df = df.reset_index().rename(columns={'index': 'SP_label'})
```

```python
df.head()
```

```python
dropped = df.drop(['SP_label', 'Total Time including setup'], axis=1, level=0)
grouped = dropped.groupby(level=0, axis=1).apply(lambda df: df.apply(lambda row: row.idxmax()[1], axis=1))
grouped.columns = pd.MultiIndex.from_product([['Best'], grouped.columns], names=['first', 'second'])
grouped
```

```python
pd.concat([df, grouped], axis=1)
```
